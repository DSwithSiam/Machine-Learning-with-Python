# 


```python
# Label encoding using for loop
import numpy as np
```

```python
for column in SuperShop_df1.columns:
    if SuperShop_df1[column].dtype == np.number:
        continue
    SuperShop_df1[column] = label.fit_transform(SuperShop_df1[column])
```

```python
# anader way with is_numeric_dtype
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
SuperShop_df1 = pd.read_csv('SuperShop.csv') 
SuperShop_df1.head(10)
```

```python
for column in SuperShop_df1.columns:
    if is_numeric_dtype(SuperShop_df1[column]):
        continue
    SuperShop_df1[column] = label.fit_transform(SuperShop_df1[column])
```
